## Setup working environment and import data

```python
import pandas as pd # Dataframes and file IO
import numpy as np # numerical calculations
```

### Import individual Hedges g values for each study, as mean for each MGI symbol

```python
df1=pd.read_csv('input_files/MoEx_forIndexHedges.csv',index_col=[1])
df1g =df1.groupby(df1.index).mean()
df1g.tail() 
```

```python
df1g.shape
```

```python
df2=pd.read_csv('input_files/MoGene_forIndexHedges.csv',index_col=[1])
df2g =df2.groupby(df2.index).mean()
df2g.tail()
```

```python
df2g.shape
```

```python
df3=pd.read_csv('input_files/RNA_Seq_forIndexHedges.csv',index_col=[1])
df3g =df3.groupby(df3.index).mean()
df3g.head()
```

```python
df3g.shape
```

```python
df4=pd.read_csv('input_files/430V2_SymbolforIndexHedges.csv',index_col=[0])
df4g =df4.groupby(df4.index).mean()
df4g.head() #shape 
```

```python
df4g.shape #
```

```python
df5=pd.read_csv('input_files/430AV2_SymbolforIndexHedges.csv',index_col=[0])
df5g =df5.groupby(df5.index).mean()
df5g.head()
```

```python
df5g.shape 
```

### Then combine these files using MGI symbol as the index , followed by a check for duplicated entries

```python
dfB = pd.concat([df1g,df2g,df3g,df4g,df5g], axis=1, join='outer') 
```

```python
dfB[dfB.index.duplicated()==True]    # Therefore no duplicate entries on index
```

```python
dfB.loc['Avp']  # have check for a familiar gene symbol
```

### Additional ENSMBL GIDs and descriptions are dropped on merging data, so import and add this data again

```python
dfX=pd.read_csv('../BioMart_Ensmbl_index/mart_export72_allMGIsymbol.txt',index_col=[3]) 
dfX.head() 
```

```python
dfX.shape
```

```python
dfX[dfX.index.duplicated()==True]  #showing duplicate entries exist in the key
```

```python
dfB.shape
```

```python
df_FINAL = dfB.join(dfX, how='left', sort=True)
df_FINAL[df_FINAL.index.duplicated()==True]  #showing duplicate entries still exist (or use .shape for number)
```

```python
dfOUT =df_FINAL.groupby(df_FINAL.index).mean()

dfOUT.shape   #check number of individual GIDs
```
